```python
from transformers import pipeline

def classify_text(email):
    classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
    labels = ["spam", "not spam"]
    hypothesis_template = "This email is {}."
    results = classifier(email, labels, hypothesis_template=hypothesis_template)

    return results['labels'][0]

    
```

```python
classify_text("You have won a free trip to the Bahamas! Click here to claim your prize!") # should return 'spam'
```

```python
classify_text('hi I am spam') # should return 'spam'
```

```python
classify_text('This is a egent mail. Click here to claim your discount') # should return 'not spam'
```
